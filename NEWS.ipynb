```python
import pandas as pd
import numpy as np
from sklearn.linear_model import PassiveAggressiveClassifier
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text  import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv("/content/drive/MyDrive/news.csv")
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data['text']=data['text'].astype(str)
```

```python
data['label']=data['label'].apply(lambda x:0 if x=='FAKE' else 1)
#or   data['label']=np.where(data['label']=='REAL',1,0)
data
```

```python
data['label'].unique()
```

```python
def rem_pun(text):
  import string
  remover=str.maketrans('','',string.punctuation)
  return text.translate(remover)
```

```python
data['text']=data['text'].apply(rem_pun)
data.head()
```

```python
nltk.download('stopwords')
```

```python
stop_words=stopwords.words("english")
```

```python
def rem_stop(text):
  text=[word.lower() for word in text.split() if word.lower() not in stop_words]
  return " ".join(text)
```

```python
data['text']=data['text'].apply(rem_stop)
data.head()
```

```python
x=data['text']
y=data['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=1)
```

```python
tfid=TfidfVectorizer()
tfid_x_train=tfid.fit_transform(x_train)
tfid_x_test=tfid.transform(x_test)

```

```python
pac=PassiveAggressiveClassifier()
pac.fit(tfid_x_train,y_train)
```

```python
pred=pac.predict(tfid_x_test)
```

```python
acc=accuracy_score(y_test,pred)
acc
```
